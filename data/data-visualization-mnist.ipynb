# Visualisation des données MNIST

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load training data
train = pd.read_csv('train.csv')
train
```

```python
train.describe()
```

### a) Afficher une image
Afficher une image quelconque depuis le train en indiquant comme titre de l'image son
label (càd le target associé) ?

```python
# random line
random_index = np.random.randint(0, len(train))
random_image = train.iloc[random_index]

# Extract label
label = random_image['label']

# Extract pixels image and convert them to matrice 28x28
pixels = random_image.drop('label').values.reshape(28, 28)

# show image with label as title
plt.imshow(pixels, cmap='gray')
plt.title(f'Label: {label}')
plt.axis('off')  # Optionnel: pour masquer les axes
plt.show()
```

### b) Afficher chaque chiffre

Afficher dans une même figure les chiffres de 0 à 9 ?

```python
# create a fig pour afficher les images
fig, axes = plt.subplots(2, 5, figsize=(10, 5))

for i in range(10):
    # select lignes correspond to number "i"
    subset = train[train['label'] == i]
    
    # select random line
    random_index = np.random.randint(0, len(subset))
    random_image = subset.iloc[random_index]
    
    # extract pixels of image and convert them to an matrice 28x28
    pixels = random_image.drop('label').values.reshape(28, 28)
    
    # Determine position axe in the grid
    ax = axes[i // 5, i % 5]
    
    # show image with label as title
    ax.imshow(pixels, cmap='gray')
    ax.set_title(f'Label: {i}')
    ax.axis('off')

plt.tight_layout()
plt.show()
```

### c) Afficher plusieurs fois le même chiffre
Pour visualiser les différentes façons d'écrire un 7, afficher les 9 1ères
images qui correspondent au chiffre 7 en les plaçant dans une même figure ?

```python
# Filter lignes correspond at number 7
sevens = train[train['label'] == 7].head(9)

# create an figure to show images
fig, axes = plt.subplots(3, 3, figsize=(8, 8))

for i, ax in enumerate(axes.flat):
    # Extract pixels of image and convert them to an matrice 28x28
    pixels = sevens.iloc[i].drop('label').values.reshape(28, 28)
    
    # show image
    ax.imshow(pixels, cmap='gray')
    ax.set_title(f'Image {i+1}')
    ax.axis('off')  # Optionnel: pour masquer les axes

# Adjust spacing beetween les subgraphs
plt.tight_layout()
plt.show()
```

### d) Représentant moyen

Afficher le représentant "moyen" de chaque chiffre ?

```python
# init an figure to show images mean
plt.figure(figsize=(20, 2))

for digit in range(10):
    # filter data for number and calculate pixels mean
    mean = train[train['label'] == digit].drop('label', axis=1).mean().values.reshape(28, 28)
    
    # show image mean
    plt.subplot(1, 10, digit + 1)
    plt.imshow(mean, cmap='gray')
    plt.title(f'Average {digit}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
